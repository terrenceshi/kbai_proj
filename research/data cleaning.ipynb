```python
import pandas as pd
import requests
import json
import spacy
import operator
import ast
```

```python
df3 = pd.read_csv('../data/imbd_data.csv')
df3.head()
```

```python
key = ""
```

```python
def getTmdbId(row):
    imdb_id = row['path'].split('/')[2]

    url = "https://api.themoviedb.org/3/find/" + imdb_id + "?api_key=" + key + "&external_source=imdb_id"

    x = requests.get(url)
    
    body = json.loads(x.text)['movie_results']
    
    if len(body) == 0:
        return -1
    
    return body[0]['id']
```

```python
df3['id'] = df3.apply(getTmdbId, axis = 1)
```

```python
def getRuntime(row):
    idd = row['id']

    url = "https://api.themoviedb.org/3/movie/"+ str(idd) +"?api_key=" + key

    x = requests.get(url)
    
    details = json.loads(x.text)
    
    if 'runtime' in details:
        rt = details['runtime']
    else:
        return -1
    
    return rt
```

```python
df4['Run Time'] = df4.apply(getRuntime, axis = 1)
```

```python
nlp = spacy.load("en_core_web_sm")

nouns = {}

def getNouns(row):
    ov = row['Overview']

    doc = nlp(ov)

    output = {}
    
    outStr = ""

    for token in doc:
        if token.pos_ == 'NOUN':
            if token.text not in output:
                output[token.text] = 1
                outStr = outStr + "," + token.text
                
            if token.text not in nouns:
                nouns[token.text] = 1
            else:
                nouns[token.text] += 1

    return outStr[1:]
```

```python
df5['nouns'] = df5.apply(getNouns, axis = 1)
```

```python
def convertYear(row):
    year = row['year']
    
    if isinstance(year, int) or isinstance(year, float):
        return int(year)
    
    if year[0] == '-':
        year = year[1:]
        
    if '(' in year:
        if '2' not in year and '1' not in year:
            return -1
        year = year.split(") (")[1][:-1]
        
    return int(year)
```

```python
df5['year'] = df5.apply(convertYear, axis = 1)
```

```python
def categorizeYear(row):
    year = row['year']
    if year >= 2010:
        return 'new'
    elif year >= 1990:
        return 'contemporary'
    else:
        return 'old'
```
