```python
def getPoster(row):
    idd = row['id']

    url = "https://api.themoviedb.org/3/movie/"+ str(idd) +"?api_key=" + key

    x = requests.get(url)
    
    details = json.loads(x.text)
    
    purl = "https://image.tmdb.org/t/p/original"
    
    if 'poster_path' in details and details['poster_path'] != None:
        output = purl + details['poster_path']
    else:
        return -1
    
    return output
```

```python
df['poster_url'] = df.apply(getPoster, axis = 1)
```
